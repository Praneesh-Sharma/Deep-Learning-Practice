```python
import torch
```

```python
#create 2 tensors
u = torch.tensor([1.0, 2.0])
v = torch.tensor([3.0, 4.0])

print(u)
print(v)
```

```python
#addition and subtraction
print(u+v)
print(u-v)
```

```python
#some common tensor functions
a = torch.rand(2,4)*2 - 1 
print(a)
print('Common functions:')
print(torch.abs(a))
print(torch.ceil(a))
print(torch.floor(a))
print(torch.clamp(a, -0.5, 0.5))
```

Major categories of Operations

    Common Functions - abs, ceil, floor, clamp, trigonometric
    Inverses of common functions - pi, sin, asin, bitwise
    Comparisions
    Reductions - max, min, std, prod, unique, vector/Matrices
    Linear algebra operations

```python
#reduction function using mean, min, max
x = torch.tensor([1, 2, 3, 4, 5], dtype=torch.float32)
print(x)
print(x.mean())
print(x.max())
print(x.min())
```

```python
#trignometric functions and their inverses
import numpy as np
x = torch.tensor([0, np.pi/2, np.pi])
print(x)
print(torch.sin(x))
print(torch.cos(x))
```

```python
#an evenly spaced list of numbers between a range
pi = torch.linspace(-np.pi/2, np.pi/2, steps=1000)
print(pi[:5]) #upper bound
print(pi[-5:]) #lower bound

sined = torch.sin(pi)
cosed = torch.cos(pi)
print(sined[0:5])
print(cosed[0:5])
```

```python
#matplotlib visualization
import matplotlib.pyplot as plt

plt.plot(sined, label='sined')
plt.plot(cosed, label='cosed')
plt.legend()
plt.show()
```
